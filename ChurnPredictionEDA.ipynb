<h2>Data Cleaning</h2>

> <h4>Dealing with missing values</h4>



```python
telcom_data=df.copy()
```

```python
#Converting TotalCharges to numerical datatype
telcom_data['TotalCharges']=pd.to_numeric(telcom_data['TotalCharges'], errors='coerce')
telcom_data.isnull().sum()
```

<h4>We can observe there are 11 missing values in TotalCharges column. Displaying the records with missing values.</h4>

```python
telcom_data.loc[telcom_data['TotalCharges'].isnull()==True]
```

```python
#Removing missing values
telcom_data.dropna(how='any',inplace=True)
```

<h4> Divide customers into bins based on tenure e.g. for tenure < 12 months: assign a tenure group if 1-12, for tenure between 1 to 2 Yrs, tenure group of 13-24; so on...

</h4>

```python
#Max tenure
print(telcom_data['tenure'].max())
```

```python
labels=["{0} - {1}".format(i,i+11) for i in range(1,72,12)]
telcom_data['Tenure_group']= pd.cut(telcom_data.tenure, range(1,80,12), right=False, labels=labels)
```

```python
telcom_data['Tenure_group'].value_counts()
```

<h4>Removing columns that are not required for processing</h4>

```python
telcom_data.drop(columns=['customerID','tenure'], axis=1 , inplace=True)
telcom_data.head()
```

<h3>Univariate Analysis</h3>

```python
for i, predictor in enumerate(telcom_data.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=telcom_data, x=predictor, hue='Churn')
```

```python
#Converting the variable 'Churn' in a binary numeric variable i.e. Yes= 1, No=0
telcom_data['Churn']=np.where(telcom_data.Churn=='Yes',1,0)
```

```python
telcom_data.head()
```

```python
#Converting categorical variables into dummy variables
telcom_data_dummies=pd.get_dummies(telcom_data)
telcom_data_dummies.head()
```

```python
#Relationship between monthly charges and total charges
sns.lmplot(data=telcom_data_dummies, x='MonthlyCharges',y='TotalCharges', fit_reg=False)
```

**BIVARIATE ANALYSIS**

```python
new_df1_target0=telcom_data.loc[telcom_data["Churn"]==0]
new_df1_target1=telcom_data.loc[telcom_data["Churn"]==1]
```

```python
def uniplot(dataframe,col,title,hue=None):
  sns.set_style("whitegrid")
  sns.set_context('talk')
  plt.rcParams["axes.labelsize"]=20
  plt.rcParams["axes.titlesize"]=22
  plt.rcParams["axes.titlepad"]=30

  temp=pd.Series(data=hue)
  fig,ax=plt.subplots()
  width=len(dataframe[col].unique()) + 7 + 4*len(temp.unique())
  fig.set_size_inches(width,8)
  plt.xticks(rotation=45)
  plt.yscale('log')
  plt.title(title)
  ax=sns.countplot(data=dataframe, x=col,order=dataframe[col].value_counts().index,hue=hue,palette='bright')

  plt.show()
```

```python
uniplot(new_df1_target1,col='Partner', title='Distribution of Gender for Churned Customers', hue='gender')
```

```python
uniplot(new_df1_target0,col='Partner', title='Distribution of Gender for Non-Churned Customers', hue='gender')
```

```python
uniplot(new_df1_target1,col='Contract', title='Distribution of Contract for Churned Customers', hue='gender')
```

```python
uniplot(new_df1_target0,col='Contract', title='Distribution of Contract for Non-Churned Customers', hue='gender')
```

```python
uniplot(new_df1_target1,col='PaymentMethod', title='Distribution of PaymentMethod for Churned Customers', hue='gender')
```

```python
uniplot(new_df1_target0,col='Contract', title='Distribution of Contract for Non-Churned Customers', hue='gender')
```

```python
uniplot(new_df1_target1,col='TechSupport', title='Distribution of TechSupport for Churned Customers', hue='gender')
```

```python
uniplot(new_df1_target1,col='SeniorCitizen', title='Distribution of Senior Citizen for Churned Customers', hue='gender')
```
